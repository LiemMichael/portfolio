```python

#!/usr/bin/env python3

from ipywidgets import VBox, HBox, Button, Output, Layout
from IPython.display import display, HTML






# Apply a dark theme to the notebook by injecting CSS.
# Uses display and HTML already imported in the notebook (from cell 0).
dark_css = """
<style>
/* general background and text */
body, .container, .notebook_app, .cell, .jp-Notebook {
    background-color: #0f1115 !important;
    color: #d6d6d6 !important;
}

/* inputs, outputs, and code cells */
div.input_area, .input, .CodeMirror, .cm-s-ipython, .CodeMirror-gutters {
    background: #0f1115 !important;
    color: #d6d6d6 !important;
    border-color: #222 !important;
}

/* rendered markdown and outputs */
.rendered_html, .output_area, .output_subarea, .output {
    background: transparent !important;
    color: #d6d6d6 !important;
    border-color: #222 !important;
}

/* prompts, metadata */
div.prompt, .prompt, .output_prompt {
    color: #9aa7b2 !important;
}

/* links and accents */
a, .rendered_html a {
    color: #8ab4ff !important;
}

/* ipywidgets */
.widget-area, .widget-box, .widget-label, .widget-hbox, .widget-vbox {
    background: transparent !important;
    color: #d6d6d6 !important;
}
.widget-button button, .btn, .button {
    background: #20242a !important;
    color: #e6e6e6 !important;
    border-color: #333 !important;
}

/* tables and borders */
table, th, td {
    border-color: #2a2a2a !important;
    color: #d6d6d6 !important;
}

/* keep code selection readable */
.CodeMirror-selected { background: rgba(138,180,255,0.15) !important; }
</style>
"""

display(HTML(dark_css))




























# --- Define your projects ---
projects = {
    "Project A": "the title of the projec comes here",
    "Project B": "the title of the projec comes here",
    "Project C": "the title of the projec comes here",
    "Project D": "the title of the projec comes here"
}

# --- Create buttons (left side) ---
buttons = []
for name in projects.keys():
    b = Button(
        description=name,
        layout=Layout(width="100%", height="60px"),
        button_style=''
    )
    buttons.append(b)

# --- Create output area (right side) ---
details = Output()

# --- Define what happens when clicking a button ---
def on_click(b):
    details.clear_output()
    with details:
        display(HTML(f"""
        <h3>{b.description}</h3>
        <p>{projects[b.description]}</p>
        <hr>
        <p><em>More details coming soon...</em></p>
        """))

for b in buttons:
    b.on_click(on_click)

# --- Layout setup ---
left_column = VBox(buttons, layout=Layout(width="30%", border="1px solid gray", padding="10px"))
right_column = VBox([details], layout=Layout(width="70%", border="1px solid gray", padding="10px"))

ui = HBox([left_column, right_column], layout=Layout(width="100%", height="400px"))
display(ui)
```
